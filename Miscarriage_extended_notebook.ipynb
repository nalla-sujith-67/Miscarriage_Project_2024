```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("Miscarriage/ No Miscarriage",axis = 1)
y = df["Miscarriage/ No Miscarriage"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,stratify=y,random_state=42)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

def model_evaluation(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    
    
    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1 Score:", f1)   
    return accuracy, precision, recall, f1
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

model = LogisticRegression(max_iter = 500)

log_accuracy = model_evaluation(model,X_train,y_train,X_test,y_test)
print(log_accuracy)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=42)
decision_tree_accuracy = model_evaluation(model,X_train,y_train,X_test,y_test)
print(decision_tree_accuracy)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
nb_accuracy = model_evaluation(model,X_train,y_train,X_test,y_test)

print(nb_accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
result = model_evaluation(rf_model,X_train,y_train,X_test,y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbm_model = GradientBoostingClassifier(n_estimators=100, random_state=42)

result = model_evaluation(gbm_model, X_train, y_train, X_test, y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier

adaboost_model = AdaBoostClassifier(n_estimators=100, random_state=42)

result = model_evaluation(adaboost_model, X_train, y_train, X_test, y_test)
```

### Cross Validation models

```python
from sklearn.model_selection import cross_val_score

log_reg_model = LogisticRegression(max_iter=1000, random_state=42)

k = 5
cv_scores = cross_val_score(log_reg_model, X_train, y_train, cv=k, scoring='accuracy')

print(f"Accuracy scores for each fold: {cv_scores}")
print(f"Mean Accuracy: {np.mean(cv_scores):.4f}")
print(f"Standard Deviation: {np.std(cv_scores):.4f}")
```

```python

dt_model = DecisionTreeClassifier(random_state=42)

k = 5
cv_scores = cross_val_score(dt_model, X_train, y_train, cv=k, scoring='accuracy')

print(f"Accuracy scores for each fold: {cv_scores}")

print(f"Mean Accuracy: {np.mean(cv_scores):.4f}")
print(f"Standard Deviation: {np.std(cv_scores):.4f}")
```

```python
nb_model = GaussianNB()
k = 5
cv_scores = cross_val_score(nb_model, X_train, y_train, cv=k, scoring='accuracy')

print(f"Accuracy scores for each fold: {cv_scores}")

print(f"Mean Accuracy: {np.mean(cv_scores):.4f}")
print(f"Standard Deviation: {np.std(cv_scores):.4f}")
```

```python
rf_model = RandomForestClassifier(random_state=42)

k = 5
cv_scores = cross_val_score(rf_model, X_train, y_train, cv=k, scoring='accuracy')

print(f"Accuracy scores for each fold: {cv_scores}")

print(f"Mean Accuracy: {np.mean(cv_scores):.4f}")
print(f"Standard Deviation: {np.std(cv_scores):.4f}")
```
